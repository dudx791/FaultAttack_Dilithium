```python
#maskkey
def get_maskedkey():
    filename = "data.bin"
    key_list = []
    # åˆå§‹åŒ–åˆ—è¡¨ä»¥å­˜å‚¨æ•°æ®
    data_list = []
    
    # å°è¯•æ‰“å¼€æ–‡ä»¶å¹¶è¯»å–æ•°æ®
    try:
        with open(filename, "rb") as file:
            # è¯»å–æ–‡ä»¶å†…å®¹
            byte = file.read(1)  # æ¯æ¬¡è¯»å–ä¸€ä¸ªå­—èŠ‚
            while byte:
                # å°†è¯»å–çš„å­—èŠ‚è½¬æ¢ä¸ºæ•´æ•°ï¼Œå¹¶æ·»åŠ åˆ°åˆ—è¡¨
                data_list.append(int.from_bytes(byte, byteorder='big'))
                byte = file.read(1)  # ç»§ç»­è¯»å–ä¸‹ä¸€ä¸ªå­—èŠ‚
    
    except FileNotFoundError:
        print("æ— æ³•æ‰“å¼€æ–‡ä»¶è¯»å–æ•°æ®ã€‚")
        exit()
    
    # è¾“å‡ºè¯»å–çš„æ•°æ®åˆ—è¡¨
    key_list = data_list[-96:]
    return key_list
print(get_maskedkey())
key_list = get_maskedkey()
keystr = ''
for item in key_list:
    key_str = hex(item)[2:].upper()
    if len(key_str) == 1:
        key_str='0' + key_str
    keystr+=key_str
print(keystr)
print(len(keystr))
```

```python
#å¿…é ˆé‹è¡Œ4
def get_input(mess_list,tr_list,key_list):
    input_list = []
    for i in range(33):
        input_list.append(mess_list[i])
    input_list.append(0)
    input_list.append(0)
    for i in range(48):
        input_list.append(tr_list[i])
    input_list.append(0)
    input_list.append(0)
    for i in range(96):
        input_list.append(key_list[i])
    input_list.append(0)
    input_list.append(0)
    return input_list
```

```python
#é€šéæ©ç¢¼å€¼è¨ˆç®—å‡ºåŸæœ¬çš„Ï'
def get_rhoprime(p_list):
    r_list = []
    for i in range(52):
        r_list.append(0)
    for i in range(52):
        r_list[i] ^= p_list[i]
        r_list[i] ^= p_list[i+64]
        r_list[i] ^= p_list[i+128]
    return r_list    
```
